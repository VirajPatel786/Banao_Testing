```python
#flickr.blogs.postPhoto
#1.Getting a Request Token
RT_Url = 'https://www.flickr.com/services/oauth/request_token/'
oauth = OAuth1(API_key, client_secret = API_secret, callback_uri = 'http://example.com')
r = requests.get(url = RT_Url, auth=oauth)
r
```

```python
#This Is Request Token 
credentials = parse_qs(r.content.decode())
OAuth_token = credentials.get('oauth_token')[0]
OAuth_secret = credentials.get('oauth_token_secret')[0]
print('OAuth Token:',OAuth_token,'\nOAuth_secret:',OAuth_secret)
```

```python
#2.Getting the User Authorization
UA_Url = 'https://www.flickr.com/services/oauth/authorize/?'
#This URl Needs To Be Entered Manually And Then Authorite It To Get The OAuth Verifer From The Redirected URL
UA_Url+urlencode({'oauth_token': OAuth_token, 'perms': 'write'}) 
```

```python
#3.Exchanging the Request Token for an Access Token
OAuth_verifier = 'cf21e183c1f22cda' #Get From Above URl
AT_Url = 'https://www.flickr.com/services/oauth/access_token/'
oauth = OAuth1(API_key, client_secret = API_secret, 
               callback_uri = 'http://example.com', 
               resource_owner_key = OAuth_token,
               resource_owner_secret = OAuth_secret,
               verifier = OAuth_verifier  )
r = requests.get(url = AT_Url, auth=oauth)
r
```

```python
r.text
```

```python
#This Is The OAuth Token !!
credentials = parse_qs(r.content.decode())
OAuth_token = credentials.get('oauth_token')[0]
OAuth_secret = credentials.get('oauth_token_secret')[0]
print('OAuth Token:',OAuth_token,'\nOAuth_secret:',OAuth_secret)
```
